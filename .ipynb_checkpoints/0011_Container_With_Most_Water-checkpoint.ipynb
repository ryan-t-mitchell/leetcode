```python
class Solution(object):
    def maxArea(self, height):
        """
        :type height: List[int]
        :rtype: int
        """
        
        #Use two pointers, head and tail:
        left = 0
        right = len(height) - 1 
        area = 0
        
        while left != right:
            distance = right - left
            
            if height[left] < height[right]:
                area = max(area, distance*height[left])
                left+=1
            else:
                area = max(area, distance*height[right])
                right-=1
        
        return area
```

```python
a = Solution()
```

```python
a.maxArea([1,8,6,2,5,4,8,3,7])
```
