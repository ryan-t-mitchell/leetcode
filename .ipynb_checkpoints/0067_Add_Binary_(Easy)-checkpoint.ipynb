Given two binary strings, return their sum (also a binary string).

The input strings are both non-empty and contains only characters 1 or 0.

Example 1:

Input: a = "11", b = "1"
Output: "100"
Example 2:

Input: a = "1010", b = "1011"
Output: "10101"
 

Constraints:

Each string consists only of '0' or '1' characters.
1 <= a.length, b.length <= 10^4
Each string is either "0" or doesn't contain any leading zero.

```python
class Solution:
    def addBinary(self, a, b):
        # In binary system, values can only be = 0 or = 1. At 0, adding 1 gives you 1. Adding again to 1 will take you back
        # to 0, but with a carry of 1. Explained here: https://www.youtube.com/watch?v=bHqLTN_jm1o
        
        # When adding, start with the last number in each string
        
        out = ''
        i = 1
        carry = 0 
        
        while i <= max(len(a), len(b)):
            try:
                if int(a[-i]) + int(b[-i]) + carry == 3:
                    carry = 1
                    out += '1'
                elif int(a[-i]) + int(b[-i]) + carry == 2:
                    carry = 1
                    out += '0'
                elif int(a[-i]) + int(b[-i]) + carry == 1:
                    carry = 0
                    out += '1'
                else: 
                    carry = 0 
                    out += '0'
            
            except:
                if len(a) > len(b):
                    if int(a[-i]) + carry == 2:
                        carry = 1
                        out += '0'
                    elif int(a[-i]) + carry == 1:
                        carry = 0
                        out += '1'
                    else:
                        carry = 0 
                        out += '0'
                else:
                    if int(b[-i]) + carry == 2:
                        carry = 1
                        out += '0'
                    elif int(b[-i]) + carry == 1:
                        carry = 0
                        out += '1'
                    else:
                        carry = 0 
                        out += '0'
            
            i += 1
        
        if carry == 1:
            out += '1'
        
        # Return the reverse of out, since we are appending each item to the end of the string in each step above
        return out[::-1]
```
